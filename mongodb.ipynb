## Create collection: VideoTranscript 

```python
video_transcript = db.create_collection('VideoTranscript')
```

```python
db.list_collection_names()
```

## The schema for VideoTranscript is as follows:

```python


# Define the schema for SingleEvent
single_event_schema = {
    "id": "se_1",
    "timestamp": datetime.datetime.now(),
    "event_type": "interaction",
    "description": "This is an example description"
}

# Define the schema for VideoEvent
video_event_schema = {
    "recording_id": "vd_1",
    "lesson_id": "ls_1",
    "created_at": datetime.datetime.now(),
    "modified_at": datetime.datetime.now(),
    "events": {
        "event_1_id": single_event_schema,
        "event_2_id": single_event_schema,
        "event_3_id": single_event_schema,
        "event_4_id": single_event_schema
    }
}

# Define the schema for VideoTranscript
video_transcript_schema = {
    "id": "vt_1",
    "video_id": "vd_1",
    "body": video_event_schema,
    "created_at": datetime.datetime.now(),
    "updated_at": datetime.datetime.now()
}

# Insert sample data into the collection
video_transcript.insert_one(video_transcript_schema)

# Define function to insert new records into VideoTranscript collection
def insert_video_transcript_record(video_id, body):
    new_record = {
        "id": f"vt_{video_transcript.count_documents({}) + 1}",
        "video_id": video_id,
        "body": body,
        "created_at": datetime.datetime.now(),
        "updated_at": datetime.datetime.now()
    }
    video_transcript.insert_one(new_record)
    return new_record

# Example usage:
new_video_transcript_record = insert_video_transcript_record("vd_2", video_event_schema)
print(new_video_transcript_record)
```

```python
video_transcript.find({"id": "vt_1"}).next()
```

## 10,000 record insertion

```python
for _ in range(1000):
    t_start = datetime.datetime.now()
    insert_video_transcript_record("vd_2", video_event_schema)
    t_end = datetime.datetime.now()
    print(f"Time taken to insert 1 records: {t_end - t_start}")
```

## Update Old Record to increment one event 

```python
def add_event_to_video_transcript(video_transcript_id, event):
    # Retrieve the existing video_transcript document
    record = video_transcript.find_one({"id": video_transcript_id})

    if record:
        # Append the new event to the events array
        record['body']['events'][f"event_{len(record['body']['events']) + 1}_id"] = event
        record['updated_at'] = datetime.datetime.now()

        # Update the document in the collection
        video_transcript.update_one({"id": video_transcript_id}, {"$set": record})
        return True
    else:
        print("Video transcript not found.")
        return False

# Example usage:
event = {
    "id": "se_2",
    "timestamp": datetime.datetime.now(),
    "event_type": "interaction",
    "description": "TuiNui"
}
add_event_to_video_transcript("vt_1", event)
```

# Update Performace Test

```python
import tqdm
video_transcript = db.VideoTranscript
t_start = datetime.datetime.now()
for _ in tqdm.tqdm(range(1000)):
    add_event_to_video_transcript("vt_1", event)
t_end = datetime.datetime.now()
print(f"Time taken to insert 1000 records: {t_end - t_start}")
# 6-7 iterations per second
```
