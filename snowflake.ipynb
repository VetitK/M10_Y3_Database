# Export Data incrementally to CSV file

```python
import os    

def determine_max_id(table_name, cur):
    cur.execute("USE SCHEMA SF_SAMPLE.LOCTH")
    cur.execute(f"SELECT MAX(ID) FROM {table_name}")
    max_id = cur.fetchone()[0]
    return max_id

def generate_csv_from_df(df, file_path, new_record_path='temp.csv', *args, **kwargs):
    if not os.path.exists(file_path):
        df.to_csv(file_path, index=False)
    else:
        existing_df = pd.read_csv(file_path)
        new_df = df[~df.index.isin(existing_df.index)]
        new_df.to_csv(file_path, mode='a', header=False, index=False)
    max_id = determine_max_id(kwargs['table_name'], kwargs['cur'])
    new_rows = []
    with open(file_path, 'r') as csv_file:
        lines = csv_file.readlines()
        for line in lines:
            if line.split(',')[0] == 'id':
                continue
            id_value = int(line.split(',')[0])  # Assuming ID is the first column
            if id_value > max_id:
                new_rows.append(line.strip())

    # Write the filtered rows to a temporary CSV file
    with open(new_record_path, 'w') as temp_csv_file:
        temp_csv_file.write("id,payment_date,course_instance_id,amount\n")
        
    with open(new_record_path, 'a+') as temp_csv_file:
        temp_csv_file.write('\n'.join(new_rows))
    return True

    
```

```python
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime  
engine = create_engine('postgresql+psycopg2://locthadmin:password@database-1.c32ikym4q9dx.ap-southeast-1.rds.amazonaws.com/postgres', echo=True, connect_args={'options': '-csearch_path={}'.format('locth')})
conn = engine.connect()
snow_curs = connect_snowflake()

def make_payment_in_sql(payment_object: dict, conn):
    # add row in payment table in sql
    payment_df = pd.DataFrame([payment_object])
    payment_df.to_sql('payment', conn, if_exists='append', index=False)
    
    payment_negative_df = payment_df.copy()
    payment_negative_df['amount'] = -payment_negative_df['amount']
    payment_negative_df['source'] = 'purchase'
    payment_negative_df['stripe_id'] = 'ch_1H9q2Z2eZvKYlo2C3sHrX5Gw'
    payment_negative_df.to_sql('payment', conn, if_exists='append', index=False)
    
    # add row in purchase table in sql
    purchase_df = pd.DataFrame()
    purchase_df['purchase_date'] = payment_df['payment_date']
    purchase_df['course_instance_id'] = payment_df['course_instance_id']
    purchase_df['student_id'] = payment_df['student_id']
    purchase_df['amount'] = payment_df['amount']
    purchase_df.to_sql('purchase', conn, if_exists='append', index=False)
    
    # add row in student_course_instance table in sql
    student_course_instance_df = pd.DataFrame()
    student_course_instance_df['student_id'] = payment_df['student_id']
    student_course_instance_df['course_instance_id'] = payment_df['course_instance_id']
    student_course_instance_df['status'] = 'Enrolled'
    student_course_instance_df.to_sql('student_course_instance', conn, if_exists='append', index=False)
    return True
    
    
def add_payment(payment_object: dict, snow_cursor, sql_connection):
    res = make_payment_in_sql(payment_object, sql_connection)
    if res:
        payment_df = pd.read_sql_table('payment', sql_connection, columns=['id', 'payment_date', 'course_instance_id', 'amount'])
        res = generate_csv_from_df(df=payment_df, file_path='payment_data.csv', new_record_path='temp.csv', table_name='Payment', cur=snow_cursor)
        upload_csv_to_snowflake(snow_curs=snow_cursor, csv_file_path='temp.csv', schema_name='SF_SAMPLE.LOCTH', table_name='Payment')
```
